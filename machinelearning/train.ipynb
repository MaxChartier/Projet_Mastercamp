```python
from pathlib import Path
import os 
from ultralytics import YOLO
```

```python
dataset = os.path.join(Path().absolute().parent, 'machinelearning', 'data')
```

```python
model = YOLO('yolo11n-cls.pt')

# Use the correct path to your dataset YAML file
model.train(task = 'classify', mode='train', data=dataset, epochs=20, imgsz=640, device='cpu')  
```

```python
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

trained_model = YOLO('best.pt')

data_test_path = os.path.join(Path().absolute().parent,'dataSet','Data','test')

# List all image files in the directory (common image extensions)
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.gif')
image_files = [f for f in os.listdir(data_test_path) if f.lower().endswith(image_extensions)]

if image_files:
    random_image = random.choice(image_files)
    random_image_path = os.path.join(data_test_path, random_image)
    print(f'Random image path: {random_image_path}')
    # Display the image using matplotlib
    img = mpimg.imread(random_image_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title(random_image)
    plt.show()
    # Predict using the model on the random image
    results = trained_model.predict(random_image_path)
    print(results)
else:
    print('No image files found in the directory.')
```
